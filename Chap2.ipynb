## if else,elif

```python
a=3  #初始化設定值
b=4
if a<=b:
    print("b是比較大的數字")  #a若小於等於b印出該行
else:   #其它結果則印出下行
    print("a是比較大的數字")
    
```

## 成績評鑑

```python
grade=87  #初始化設定值
print("成績評鑑：")
if grade>=90:    #用if、else if(elif)判斷並印出文字
    print ("優等")
elif grade>=80:
    print("甲等")
elif grade>=70:
    print("乙等")
elif grade>=60:
    print("丙等")
else:
    print("不及格")
```

# for迴圈控制

```python
#1+2+...+10總和
sum=0
for count in range(1,11,1): #從1開始，間隔1，不含11
    sum+=count #加count的值(1到10)
print("1+2+...+10=",sum)
```

```python
#range另種寫法
#1+2+...+10總和
sum=0
for count in range(11):  #從1開始，間隔1，不含11
    sum+=count #加總count的值(1到10)
print("1+2+...+10=",sum)
```

雞兔同籠，若共又35個頭，94隻腳，請問有幾隻雞、兔?

```python
for chi in range(0,36):
    rab=35-chi
    if(2*chi+4*rab)==94:
        print("雞有",chi,"隻,兔子有",rab,"隻")
        break
else:
    print("無解")
```

# while迴圈

```python
sum = 0  #初始化設定值
count=1  #初始化設定值
while count <=10:  #執行直到count不小於等於10為止
    sum += count  #加count值
    count+=1 #count每次+1
print("1+2+3+...+10=",sum)
```

### 本金10000元，定存年利率0.6%，請問多久銀行存款會翻倍

```python
x=10000  #初始化設定值
years=0  #初始化設定值
while x<=20000:  #執行直到x不小於等於20000為止
    x=x*1.006   #3計算
    years+=1 #每次+1
print(str(years),"年以後，存款加倍，為",int(x),"元")
```

### 帳號密碼判斷

```python
id="ying" #初始化設定值id=ying
pwd="1111" #初始化設定值pwd=1111
while True: #執行迴圈
    x=input("請輸入帳號:")
    y=input("請輸入密碼:")
    if(x==id and y==pwd):
        print("歡迎登入")
        break;  #停止跑迴圈
    else:
        print("帳號密碼輸入錯誤")
```

猜數字

```python
import random  #插入亂數
number=random.randint(1,20)
while True:
    guess=int(input("請輸入1~20之間的數字-->"))
    if guess == number:
        print("你猜對了，數字是", number)
        break
    else:
        print("你猜錯了")
```

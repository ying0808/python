# 函式(數)
## 傳統的寫法

```python
import math  #插入數學物件
radius = eval(input("請輸入圓的半徑：")) #把輸入的字串變成數學計算型態
print("圓的面積為：", radius * radius * math.pi) #印出值
```

## 面積計算函式

```python
import math  #插入數學物件
def circle_area(radius):  #函式
    area = radius * radius * math.pi #運算式
    return area  #回傳值

print(circle_area(10)) #將函式帶入值=10
```
